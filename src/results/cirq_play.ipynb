```python
import json
import matplotlib.pyplot as plt

with open("cirq_cnot_vs_d_32.json") as f:
    results = json.load(f)
```

```python
# plot of the unoptimized cnot counts

plt.figure()

for encoding, (ds, raw, opt) in results.items():
    plt.plot(ds, raw, marker="o", label=f"{encoding} (optimized)")

plt.xlabel("Bosonic truncation")
plt.ylabel("CNOT count")
plt.legend()
plt.show()
```

```python
# optimized cnot counts

plt.figure()

for encoding, (ds, raw, opt) in results.items():
    plt.plot(ds, opt, marker="o", label=f"{encoding} (optimized)")

plt.xlabel("Bosonic truncation")
plt.ylabel("CNOT count")
plt.legend()
plt.show()
```

```python
# unoptimized and optimized

plt.figure()

for encoding, (ds, raw, opt) in results.items():
    plt.plot(ds, raw, marker="o", label=f"{encoding} (pre-optimization)")
    plt.plot(ds, opt, marker="v", label=f"{encoding} (post-optimization)")

plt.xlabel("Bosonic truncation (d)")
plt.ylabel("CNOT count")
plt.legend()
plt.show()
```

```python
# For unary (d = 32): pre-qiskit op cnot count=124, post-qiskit op cnot count=124
# For gray (d = 32): pre-qiskit op cnot count=320, post-qiskit op cnot count=320
# For standard binary: pre-qiskit op cnot count=480, post-qiskit op cnot count=480
```
